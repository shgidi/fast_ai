```python
from scipy.special import gamma
from scipy import integrate
from matplotlib import pyplot as plt
% matplotlib inline
import numpy as np
```

```python
n=35
n_b=15
```

```python
from sympy import *
```

```python
numerator= (lambda x:x**n_b*(1-x)**(n-n_b))
```

```python
denom=integrate.quad(numerator,0,1)[0] #result and upeer error bound
```

```python
#neat!
xs=np.arange(0,1,0.01)
plt.plot(xs,list(map(numerator,xs)))
```

```python
#beta function
alp,bet=6,3
beta=lambda x:x**alp*(1-x)**bet
plt.plot(xs,list(map(beta,xs)))
```
